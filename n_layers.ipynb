```python
def init_param(l):
    ws=[]
    bs=[]
    l= l-1
    for l in range(l):
        w=np.random.rand(l+1,l)-0.5
        b=np.random.rand(l+1,1)-0.5
        ws.append(w)
        bs.append(b)
    return ws,bs    

def Relu(z):
    return np.maximum(0,z)
def softmax(z):
    A=np.exp(z)/sum(np.exp(z))
    return A
def farward_prop(ws,bs,X,l):
    zs=[]
    As=[]
    zl=[]
    Al=[]
    l=l-1
    for l,w,b in l,ws,bs:
        if l==1:
            z=w[l].dot(X)+b[l]
            A= Relu(z)
            zs.append(z)
            As.append(A)
        elif l>1 and l<-1:
            z=ws[l].dot(As[l]) + b[l]  
            A= Relu(z)
            zs.append(z)
            As.append(A)
        else:
            z= w[l].dot(As[l])+b[l]
            A=softmax(z)
            zl.append(z)
            Al.append(A)

    return zs, As,zl,Al  

def one_hot(y):
    hot=np.zeros((y.size,y.max()+1))
    hot[np.arange(y.size),y]=1
    one_hot=hot.T
    return one_hot

def Relu_Der(z):
    return z>0

def backward_prop(ws,zs,As,Al,Y,l):
    one_hot_y=one_hot(Y)  
    dzl=Al-one_hot_y
    dz=[]
    dw=[]
    db=[]
    l=l-1
    for l,w,z,A, in l,ws,zs,As:
        dz=w[l+1].T.dot(dz[l+1])*Relu_Der(z[l])    
        dw=1/m*dz[l].dot(A[l].T)
        db=1/m*np.sum(dz[l])
        dz.append(dz)
        dw.append(dw)
        db.append(db)
    return dw,dz,db

def update_params(ws,bs,dw,db,lr,l):
    W_new=[]
    B_new=[]
    for w,b,l,dw,db in ws,bs,l,dw,db:
        w=w[l]-lr*dw[l]
        b=b[l]-lr*db[l]
        W_new.append(w)
        B_new.append(b)

    return W_new,B_new    

     
```

```python
def get_predictions(A2):
    return np.argmax(A2, 0)

def get_accuracy(predictions, Y_train):
    print(predictions, Y_train)
    return np.sum(predictions == Y_train) / Y_train.size

def gradient_descent(X_train, Y_train,l, lr, iterations):
    ws, bs = init_param(l)
    for i in range(iterations):
        Zs, As, Zl, Al = farward_prop(ws,bs,X_train,l)
        dw,dz,db = backward_prop( ws,Zs, As, Al,Y_train,l)
        ws, bs = update_params(ws, bs, dw,db,lr,l)
        if i % 200 == 0:
             print("Iteration: ", i)
             predictions = get_predictions(Al)
             print(get_accuracy(predictions, Y_train))

    return ws,bs  
```
